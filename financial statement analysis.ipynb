```python
#Data
revenue = [14574.49,7606.46,8611.41,9175.41,8058.65,8105.44,11496.28,9766.09,10305.32,14379.96,10713.97,15433.50]
expenses = [12051.82,5695.07,12319.20,12089.72,8658.57,840.20,3285.73,5821.12,6976.93,16618.61,10054.37,3803.96]
```

```python
#calculating profit for each month
profit = list([])
for i in range(0,len(revenue)):
    profit.append(revenue[i] - expenses[i])
profit
```

```python
#calaculating  30% tax weigh
tax=([round(i*0.30,2) for i in profit])
```

```python
tax
```

```python
#caluculating profit after tax
profit_after_tax = list([])
for i in range(0,len(tax)):
    profit_after_tax.append(profit[i]-tax[i])
profit_after_tax
```

```python
#calculating profit margin in percentage
profit_margin = list([])
for i in range(0,len(profit_after_tax)):
    profit_margin.append(profit_after_tax[i]*100 / revenue[i])
profit_margin
```

```python
mean = sum(profit_after_tax)/len(profit_after_tax)
mean
```

```python
# good months
good_months = list([])
for i in range(0,len(profit_after_tax)):
    good_months.append(profit_after_tax[i] > mean )
good_months
```

```python
#Bad months
bad_months = list([])
for i in range(0,len(profit_after_tax)):
    bad_months.append(profit_after_tax[i] < mean )
bad_months
```

```python
#best month
best_month = list([])
for i in range(0,len(profit)):
    best_month.append((profit_after_tax[i])==max(profit_after_tax))
best_month
```

```python
#bad month
bad_month = list([])
for i in range(0,len(profit)):
    bad_month.append((profit_after_tax[i])==min(profit_after_tax))
bad_month
```
